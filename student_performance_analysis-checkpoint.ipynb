### 2.cleaning the data 

```python
data.isnull().sum() ## checkin for the null values
```

```python
###  we r dropin the race / etnicity and parent where its not required for the analysis 
```

```python
student=data.drop(['race/ethnicity','parental level of education'],axis=1)
```

```python
student.head()
```

### 3.Relatioship Analysis

### heatmap

```python
corelation=student.corr()
```

```python
sns.heatmap(corelation,xticklabels=corelation.columns,yticklabels=corelation,annot=True)
```

### pairplot = if we only want to visualize btw the 2 variables then we use pair plot 

```python
sns.pairplot(student)
```

## scatter plot  used to plot the relationship btw the 2 integers 

```python
sns.relplot(x='math score',y='writing score',hue='gender',data=student)
```

### histogram is a graphical display of bars of different heights

```python
sns.distplot(student['writing score'],bins=5)   ### bins is used for zooming 
```

### box plot 

```python
sns.catplot(x='math score',kind='box',data=student)
```

```python
sns.catplot(x='writing score',kind='box',data=student)
```
